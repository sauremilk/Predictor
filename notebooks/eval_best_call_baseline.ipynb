```python
# Zelle 3: Beispiel - Lade finale Pipeline und mache eine Vorhersage f√ºr einen neuen Game-State
import joblib
import pandas as pd
pipeline = joblib.load('models/baseline_pipeline_final.joblib')
# Beispiel-Featurezeile (als dict) - passe Werte an das Feature-Set an
example = {
    'zone_phase': 'mid',
    'alive_players': 34,
    'zone_index': 5,
    'storm_edge_dist': 12,
    'mats_total': 450,
    'surge_above': 10,
    'height_status': 'mid',
    'position_type': 'edge',
    'teammates_alive': 2,
    'outcome_placement': 20,
    'outcome_alive_time': 120
}
example_df = pd.DataFrame([example])
pred = pipeline.predict(example_df)
proba = pipeline.predict_proba(example_df)
print('Predicted best_call:', pred[0])
print('Class probabilities:', proba[0])
```
